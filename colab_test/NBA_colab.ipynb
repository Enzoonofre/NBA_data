```python
#Informações de desempenho geral, abaixo há uma descrição do que é cada atributo e abaixo disso o DataFrame que contém esses atributos
'''
W = vitórias
L = derrotas
W_PCT = porcentagem de vitórias

DivisionRank = posição do time na divisão
HOME = desempenho em casa (vitórias e derrotas)
ROAD = desempenho fora de casa (vitórias e derrotas)

ThreePTSOrLess = jogos decididos por 3 pontos ou menos
TenPTSOrMore = jogos com diferença de 10 ou mais pontos
LongWinStreak = maior sequência de vitórias
LongLossStreak = maior sequência de derrotas
PointsPG = média de pontos marcados por jogo
OppPointsPG = média de pontos sofridos por jogo
DiffPointsPG = diferença média entre pontos marcados e sofridos por jogo

vsEast = desempenho contra times da Conferência Leste
vsAtlantic = desempenho contra times da Divisão Atlantic
vsCentral = desempenho contra times da Divisão Central
vsSoutheast = desempenho contra times da Divisão Southeast
vsWest = desempenho contra times da Conferência Oeste
vsNorthwest = desempenho contra times da Divisão Northwest
vsPacific = desempenho contra times da Divisão Pacific
vsSouthwest = desempenho contra times da Divisão Southwest

Jan = desempenho nos jogos de janeiro
Feb = desempenho nos jogos de fevereiro
Mar = desempenho nos jogos de março
Apr = desempenho nos jogos de abril
May = desempenho nos jogos de maio
Jun = desempenho nos jogos de junho
Jul = desempenho nos jogos de julho
Aug = desempenho nos jogos de agosto
Sep = desempenho nos jogos de setembro
Oct = desempenho nos jogos de outubro
Nov = desempenho nos jogos de novembro
Dec = desempenho nos jogos de dezembro
'''

performance_standings = pd.DataFrame()
performance_standings = standings[['WINS','LOSSES','WinPCT','DivisionRank','HOME','ROAD','ThreePTSOrLess','TenPTSOrMore','LongWinStreak','LongLossStreak','PointsPG','OppPointsPG','DiffPointsPG','vsEast','vsAtlantic','vsCentral','vsSoutheast','vsWest','vsNorthwest','vsPacific','vsSouthwest','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']]
performance_standings.head()
```

```python
#Informações importantes de referência
reference_standings = pd.DataFrame()
reference_standings = standings[['LeagueID','SeasonID','TeamID','TeamCity','TeamName','Conference','Division']]
reference_standings.head()

```

```python
clean_standings = pd.concat([reference_standings,performance_standings],axis=1)
clean_standings.info()
```
